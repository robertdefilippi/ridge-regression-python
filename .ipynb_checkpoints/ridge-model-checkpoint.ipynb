```python
%matplotlib inline
import numpy as np
import random as random
import matplotlib.pyplot as plt

# Set seed
random.seed(1075)

# Create random curve
length = 125
x = np.linspace(0, 10, length)
y = .5 * np.cos(x)
y_scatter = []

# Create random scattering around the true function
for i, j in enumerate(y):
    y_scatter.append( y[i] + random.uniform(-.3, .3))   

plt.figure(figsize=(12, 7))    
plt.plot(x, y)
plt.scatter(x, y_scatter)
plt.title("Scatter Vs. Actual")
plt.legend(['True Function', 'Observed Points'])
```

```python
# Try fitting a simple linear function with a polynomial order
from sklearn.linear_model import LinearRegression
from numpy import vander

# We need to fit a polynomial function to our data points using the vander function
# If we just used linear regression we obtain a poor fit because our data was generated  
# using a nonlinear function. To get linear estimates of our coefficients from non-linear inputs
# we increase the number of dimensions of our data. I'm taking a guess our data is polynomial of
# degree 3, so I'm using 4 (3 + 1) as the parameter in vander to return that many columns
# If I wanted to test polynomial of degree 10, I would put 11 as the parameter.

lm_model_3 = LinearRegression()
lm_model_3.fit(vander(x, 4), y_scatter)
degree_3 = lm_model_3.coef_.size - 1
y_pred_3 = lm_model_3.predict(np.vander(x, degree_3 + 1))

lm_model_8 = LinearRegression()
lm_model_8.fit(vander(x, 9), y_scatter)
degree_8 = lm_model_8.coef_.size - 1
y_pred_8 = lm_model_8.predict(np.vander(x, degree_8 + 1))

# Plot side by size
plt.figure(figsize=(12, 7))    
plt.plot(x, y)
plt.plot(x, y_pred_3)
plt.plot(x, y_pred_8)
plt.scatter(x, y_scatter)
plt.title("Scatter Vs. Actual")
plt.legend(['True Function', 'Pred. Deg. 3', 'Pred. Deg. 8', 'Observed Points'])
```

```python
lm_model_30 = LinearRegression()
lm_model_30.fit(vander(x, 31), y_scatter)
degree_30 = lm_model_30.coef_.size - 1
y_pred_30 = lm_model_30.predict(np.vander(x, degree_30 + 1))

# Plot side by size
plt.figure(figsize=(12, 7))    
plt.plot(x, y)
plt.plot(x, y_pred_30)
plt.scatter(x, y_scatter)
plt.title("Scatter Vs. Actual")
plt.legend(['True Function', 'Pred. Deg. 30', 'Observed Points'])
```

```python
# Looks like we could be overfitting our data if we increase the polynomial degree too much
# How do we find the right polynomial number?

from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd

# Get the RMSE for each predicted values
def get_rmse(y, y_pred):
    return np.sqrt(metrics.mean_squared_error(y, y_pred))

# Create dataframe to evaluate
rmse_df = pd.DataFrame(columns=["degree", "rmse_train", "rmse_test"])

# Number of degress to test in our model
train_X, test_X, train_y, test_y = train_test_split(x, y_scatter,
                                                    test_size=0.33,
                                                    random_state=1075)

# Get the rmse for each prediction
for i in range(1, 10):
    p = np.polyfit(train_X, train_y, deg=i)
    rmse_df.loc[i-1] = [i,
                        get_rmse(train_y, np.polyval(p, train_X)),
                        get_rmse(test_y, np.polyval(p, test_X))]
```

```python
# Plot the two RMSE of the training and the test data
plt.figure(figsize=(12, 7))
plt.plot(rmse_df.degree, rmse_df.rmse_train, label='Training Data')
plt.plot(rmse_df.degree, rmse_df.rmse_test, label='Test Data')
plt.ylabel('RMSE')
plt.xlabel('Polynomial Degree')
plt.legend()
plt.title('Train Vs. Test Error')
```

```python
# It can observed our model starts to overfit ... as we would expect with a small dataset ... around 5
# Here there is no real increase or trade off between the error terms after 5
lm_model_5 = LinearRegression()
lm_model_5.fit(vander(x, 6), y_scatter)
degree_5 = lm_model_5.coef_.size - 1
y_pred_5 = lm_model_5.predict(np.vander(x, degree_5 + 1))

# Plot side by size
plt.figure(figsize=(12, 7))    
plt.plot(x, y)
plt.plot(x, y_pred_5)
plt.scatter(x, y_scatter)
plt.title("Scatter Vs. Actual")
plt.legend(['True Function', 'Pred. Deg. 5', 'Observed Points'])
```

```python
# Let's try a different set of data
def curve(x_val):
    return (x_val ** 2) + 3.0

y_scatter_curve = curve(x)
noise = 20 * (np.random.random(length) - 4.0)
y_scatter_curve = y_scatter_curve + noise

plt.figure(figsize=(12, 7)) 
plt.scatter(x, y_scatter_curve)
plt.title("Scatter Curve To Estimate")
```

```python
plt.figure(figsize=(12, 7))
plt.plot(rmse_df_curve.degree, rmse_df_curve.rmse_train, label='Training Data')
plt.plot(rmse_df_curve.degree, rmse_df_curve.rmse_test, label='Test Data')
plt.ylabel('RMSE')
plt.xlabel('Polynomial Degree')
plt.legend()
plt.title('Train Vs. Test Error')
```

```python
# Here we see a divergence ... so we'll stick to Poly level 2
# It can observed our model starts to overfit ... as we would expect .. around 5
# Here there is no real increase or trade off between the error terms after 5
# Mostly because we're extimating a cos wave, and it will be roughly the same 
lm_model_2 = LinearRegression()
lm_model_2.fit(vander(x, 3), y_scatter_curve)
degree_2 = lm_model_2.coef_.size - 1
y_pred_2 = lm_model_2.predict(np.vander(x, degree_2 + 1))

# Plot side by size
plt.figure(figsize=(12, 7))    
plt.plot(x, y_pred_2)
plt.scatter(x, y_scatter_curve)
plt.title("Scatter Vs. Actual")
plt.legend(['Pred. Deg. 2', 'Observed Points'])
```

```python
# Add some more bumps to the data

# Make a copy
y_scatter_curve_noise = y_scatter_curve.copy()

for i, j in enumerate(y_scatter_curve_noise):
    if i < 25:
        y_scatter_curve_noise[i] = j + (5 * (np.random.random() + 5.0))
    if i > 100:
        y_scatter_curve_noise[i] = j + (5 * (np.random.random() - 5.0))

plt.figure(figsize=(12, 7))    
plt.scatter(x, y_scatter_curve_noise)
plt.title("Scatter With More Noise")
```

```python
# Create graphs of the different values of alpha you're testing
from sklearn.linear_model import Ridge

# Set the expected values of lambda
alphas = np.linspace(.00001, 2, 500)
alphas_to_display = [alphas[0], alphas[100], alphas[200], alphas[300], alphas[400], alphas[499]]
ridge_error_df = pd.DataFrame(columns=["alpha", "rss", "intercept", "coef"])

def ridge_model_comparison(alphas, poly_degree, X_values, y_values):
    # Set local variables
    count = 0
    subplot = 1
    fig = plt.figure(figsize=(12, 8))
    
    # Construct your model to evaluate
    for i in alphas:
        ridge_model = Ridge(alpha=i, normalize=True)
        ridge_model.fit(vander(x, poly_degree + 1), y_values)
        ridge_degree = ridge_model.coef_.size - 1
        y_pred = ridge_model.predict(np.vander(x, ridge_degree + 1))

        # Only display certain models
        if i in alphas_to_display:
            plt.subplot(230 + subplot)
            plt.tight_layout()
            plt.plot(X_values, y_pred)
            plt.plot(X_values, y_values, '.')
            plt.title('Plot for alpha: %.3g on Poly. Deg %d ' % (i, poly_degree))
            subplot = subplot + 1

        # Fill dataframe
        rss = sum((y_pred - y_values)**2)
        intercept = ridge_model.intercept_
        coef = ridge_model.coef_

        # Add error data to the dataframe
        # alpha, rss, intercept, coef
        ridge_error_df.loc[count] = [i, rss, intercept, coef]
        count = count + 1

# Run the model
ridge_model_comparison(alphas=alphas, poly_degree=4, X_values=x, y_values=y_scatter_curve_noise);
```

```python
# Try running the model for a Lasso regression
from sklearn.linear_model import Lasso

def lasso_model_comparison(alphas, poly_degree, X_values, y_values):
    lasso_error_df = pd.DataFrame(columns=["alpha", "rss", "intercept", "coef"])
    # Set local variables
    count = 0
    subplot = 1
    fig = plt.figure(figsize=(12, 8))
    
    # Construct your model to evaluate
    for i in alphas:
        lasso_model = Lasso(alpha=i, normalize=True)
        lasso_model.fit(vander(x, poly_degree + 1), y_values)
        lasso_degree = lasso_model.coef_.size - 1
        y_pred = lasso_model.predict(np.vander(x, lasso_degree + 1))

        # Only display certain models
        if i in alphas_to_display:
            plt.subplot(230 + subplot)
            plt.tight_layout()
            plt.plot(X_values, y_pred)
            plt.plot(X_values, y_values, '.')
            plt.title('Plot for alpha: %.3g on Poly. Deg %d ' % (i, poly_degree))
            subplot = subplot + 1

        # Fill dataframe
        rss = sum((y_pred - y_values)**2)
        intercept = lasso_model.intercept_
        coef = lasso_model.coef_

        # Add error data to the dataframe
        # alpha, rss, intercept, coef
        lasso_error_df.loc[count] = [i, rss, intercept, coef]
        count = count + 1

# Run the function
lasso_model_comparison(alphas=alphas, poly_degree=4, X_values=x, y_values=y_scatter_curve_noise);
```
